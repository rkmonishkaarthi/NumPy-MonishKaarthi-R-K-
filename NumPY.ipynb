```python
import numpy as np

a = np.array([10,20,30,40])
b = np.array([40,50,60,70])

print(np.sum(a))
print(np.sum(b))
print(np.max(a))
print(np.max(b))
print(a+b)
print(a/b)
print(a+10)
print(a*3)
print(np.std(a))
print(np.std(b))
print(a%b)
print(np.mean(a))
print(np.mean(b))
```
